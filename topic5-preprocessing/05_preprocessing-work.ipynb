## Principal Component Analysis

https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

***

```python
# Create a new PCA instanece
pca = dec.PCA(n_components=2)
```

```python
# The X data
X
```

```python
# Fit the data to pca
pca.fit(X)
```

```python
# Show the variability
pca.explained_variance_ratio_
```

```python
# Transform X
X_pca = pca.transform(X)

# Show
X_pca
```

```python
# Create an empty plot
fig, ax = plt.subplots()

# Plot a scatter plot
ax.plot(X_pca[:, 0], X_pca[:, 1], 'k.')
```

```python
# Original classification
df_pca = pd.DataFrame(df[['species', 'sex']])
# Show
df_pca
```

```python
# Incorporate our PCA variables.
df_pca['pca0'] = X_pca[:, 0]
df_pca['pca1'] = X_pca[:, 1]

# Show.
df_pca
```

```python
# Pair plot
sns.pairplot(df_pca, hue = 'species')
```

```python
# The scaled data.
df_X_transformed
```

```python
# Ceate a new PCA instance
pca = dec.PCA(n_components=2)

# Fit the scaled data
pca.fit(df_X_transformed)

# Transform
X_transformed_pca = pca.transform(df_X_transformed)

# Original classification
df_trans_pca = pd.DataFrame(df[['species', 'sex']])

# Incorporate our PCA variables
df_trans_pca['pca0'] = X_transformed_pca[:, 0]
df_trans_pca['pca1'] = X_transformed_pca[:, 1]

# Show
df_trans_pca
```

```python
# Look at the variance
pca.explained_variance_
```

```python
# Pair plot
sns.pairplot(df_trans_pca, hue='species')
```

## kNN on the scaled data

***

### Before Scaling

```python
# Load the data.
df = pd.read_csv('notes/data/penguins.csv').dropna()

# Create a new instance of a classifier.
clf = sk.neighbors.KNeighborsClassifier()

# X data.
X = df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]

# y data.
y = df['species'].to_numpy()

# Run cross-validation with five folds.
sk.model_selection.cross_val_score(clf, X, y)
```

### After Scaling

```python
# Load the data.
df = pd.read_csv('notes/data/penguins.csv').dropna()

# Create a new instance of a classifier.
clf = sk.neighbors.KNeighborsClassifier()

# X data.
X = df[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]

# Use a standard scaler
scaler = pre.StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

# y data
y = df['species'].to_numpy()

# Run cross-validation with five folds.
sk.model_selection.cross_val_score(clf, X, y)
```
